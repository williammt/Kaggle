```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
videos = pd.read_csv("C:\\Users\\William.Tang\\OneDrive - Prometric Inc\\Documents\\Kaggle Datasets\\archive (2)\\List of most-viewed YouTube videos.csv")
videos.info()
```

```python
videos.head()
```

```python
videos['Upload Date'] = pd.to_datetime(videos['Publication date'])
```

```python
plt.bar(videos['Video name'],videos["Views (billions)"])
plt.xticks(videos['Video name'], rotation = "vertical")
plt.xlabel('Video Name')
plt.ylabel('Views in Billions')
```

```python
dates = [date for date, df in videos.groupby('Upload Date')]
videos = videos.sort_values(by=['Upload Date'])
```

```python
plt.bar(range(0,videos.shape[0]),videos["Views (billions)"])
plt.xticks(range(0,videos.shape[0]), videos['Upload Date'],rotation = 'vertical')
plt.xlabel('Video Date')
plt.ylabel('Views in Billions')
# top 3 were made 2nd half of 2016 to start of 2017
```
